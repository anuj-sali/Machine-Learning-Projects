```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
df=pd.read_csv("train.csv")
df.dropna(inplace=True)
df_Xtrain=pd.get_dummies(df,columns=['PRIM_STATE','NAICS_TITLE','OCC_TITLE'])
df1=df_Xtrain.copy()
df1.drop(['A_PCT10','A_PCT25','A_MEDIAN','A_PCT75','A_PCT90'],axis=1,inplace=True)
X=df1
y=df[['A_PCT10','A_PCT25','A_MEDIAN','A_PCT75','A_PCT90']]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)

# define model
rf_model = RandomForestRegressor(random_state=0)
# fit model
rf_model.fit(X_train,y_train)
train_rf_score = rf_model.score(X_train,y_train)
test_rf_score = rf_model.score(X_test,y_test)
y_pred = rf_model.predict(X_test)
```

```python
#train,test score (r2 score)
train_rf_score,test_rf_score
```

```python
df_pred=pd.DataFrame(y_pred)
y_test=y_test.reset_index(drop=True)
df_concat=pd.concat([y_test,df_pred],axis = 1)
df_concat.head()
```

```python
def MNWAE(df):
    df["10_"]=df.apply(lambda x :  0.1*(abs(x[5] - x[0])/x[0]) ,axis=1)
    df["25_"]=df.apply(lambda x :  0.2*(abs(x[6] - x[1])/x[1]) ,axis=1)
    df["50_"]=df.apply(lambda x :  0.4*(abs(x[7] - x[2])/x[2]) ,axis=1)
    df["75_"]=df.apply(lambda x :  0.2*(abs(x[8] - x[3])/x[3]) ,axis=1)
    df["90_"]=df.apply(lambda x :  0.1*(abs(x[9] - x[4])/x[4]) ,axis=1)
    mnwae=(df[[ '10_','25_', '50_', '75_', '90_']].sum().sum())/ len(df)
    return mnwae
```
